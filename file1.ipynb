```python
import pandas as pd
```

```python

airport = pd.read_csv(r'data/airports.csv')
airport1 = pd.read_csv(r'data/us-airports.csv')
birdstrike = pd.read_csv(r'data/birdstrikes.csv')
```

```python

airport_data = pd.DataFrame(birdstrike.drop_duplicates(subset=['Airport Name'], keep='first').reset_index())
airport_data['Airport Name'] = airport_data['Airport Name'].str.replace(' ARPT', ' AIRPORT')
airport_data['Airport Name'] = airport_data['Airport Name'].str.replace(' INTL', ' INTERNATIONAL')
airport_data = pd.DataFrame(airport_data[['Airport Name','Origin State']])
airport_data.rename(columns={'Origin State':'state'}, inplace=True)
airport_data.rename(columns={'Airport Name':'airport'}, inplace=True)
```

```python
airport['AIRPORT'] = airport['AIRPORT'].str.upper()
air = pd.DataFrame(airport[['AIRPORT', 'LATITUDE', 'LONGITUDE']])
```

```python
airport1['name'] = airport1['name'].str.upper()
air1 = pd.DataFrame(airport1[['name', 'latitude_deg', 'longitude_deg']])
airport_data['latitude'] = 0
airport_data['longitude'] = 0
```

```python
x = airport_data['airport']
y = air['AIRPORT']
lt = air['LATITUDE']
lg = air['LONGITUDE']
a_lt = airport_data['latitude']
a_lg = airport_data['longitude']

for i in range(len(x)):
    for j in range(len(y)):
        if x[i] == y[j]:
            a_lt[i] = lt[j]
            a_lg[i] = lg[j]
airport_data['latitude'] = a_lt
airport_data['longitude'] = a_lg
```

```python
import csv
import json

class Node:
    def __init__(self, name, size=0):
        self.name = name
        self.size = size
        self.children = {}

    def __getitem__(self, idx):
        if idx not in self.children:
            self.children[idx] = Node(idx)
        return self.children[idx]

    def as_dict(self):
        res = { "name": self.name }
        if self.size:
            res["size"] =  self.size
        else:
            res["children"] = [v.as_dict() for v in self.children.values()]
        return res

root = Node('Bird Strikes')

with open('data/birdstrikes_to_json.csv', 'r') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        # grp1, grp2, grp3, size = row
        # root[grp1][grp2][grp3].size += size
        root[row[2]][row[0]][row[3]][row[4]].size += 1

print (json.dumps(root.as_dict(), indent=5))
jsonFilename = r'data/Heirarchy.json'
with open(jsonFilename, 'w') as f2:
    (json.dump(root.as_dict(), f2, indent=4,ensure_ascii=False))
```
